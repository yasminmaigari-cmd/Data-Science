```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('dark_background')
```

1. make a trigonometric array


```python
x = np.linspace(0,360,100)
xrad = np.deg2rad(x)

y1=np.sin(xrad)
y2=np.cos(xrad)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot('trigonometric Function')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
trig = pd.DataFrame({'X': x, 'sin(x)': y1, 'cos(x)' :y2})

trig.tail(10)
```
